### Расчет эффективности обогащения.

Создадим функцию, расчитывающую эффективность процесса обогощения золотосодержащей руды. Для этого воспользуемся формулой: 
$$Recovery = \frac{C*(F-T)}{F*(C-T)}* 100\% $$ где:

* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

```python
def recovery(feed_in, concentrate_out, tail_out):
    rec = concentrate_out * (feed_in - tail_out) / \
          (feed_in * (concentrate_out - tail_out)) * 100
    return rec
```

Проверим коректность формулы на обучающих данных. В качестве критерия оценки возьмем sMAPE (симметричное среднее абсолютное процентное отклонение). Для этого воспользуемся формулой: 
$$sMAPE = \frac{1}{N} \sum_{ i=1 }^{ N } \frac{|y_i - \hat{y_i}|}{|y_i| + |\hat{y_i}|/2}* 100\% $$ где:

* N — количество элементов в выборке;
* y_i — значение целевого признака для i-того элемента;
* \hat{y_i} — значение предсказания для i-того элемента.

```python
def s_mape(target, prediction):
   
    numerator = abs(target - prediction)
    denominator = (abs(target) + abs(prediction))/2
   
    return np.mean(numerator / denominator)* 100
```

```python
prediction_test = pd.Series(recovery(data_train['rougher.input.feed_au'],
                            data_train['rougher.output.concentrate_au'],
                            data_train['rougher.output.tail_au']))
```

```python
mae = mean_absolute_error(data_train['rougher.output.recovery'], prediction_test)

print('Cредняя абсолютная ошибка на обучающих данных равна', mae, '.')
```

## Модель

Теперь можем перйти к созданию предсказывающей модели. Модели нужно спрогнозировать сразу две величины - эффективность обогащения чернового концентрата(rougher.output.recovery) и эффективность обогащения финального концентрата(final.output.recovery).

Для этого проведем следующие исследования:
- создадим функцию для вычисления итоговой sMAPE
- обучим несколько моделей и оценим их качество кросс-валидацией
- выберем лучшую моодель и проведем проверку на тестовых данных


### Функция для вычисления итоговой sMAPE.

Так как модель прогнозирует сразу 2 признака, внесем корректировку в расчёт критерия оценки sMAPE (симметричное среднее абсолютное процентное отклонение). Теперь он будет расчитываться по следующей формуле:
$$sMAPE = 25\% * sMAPE(rougher) + 75\% * sMAPE(final)$$ где:

* sMAPE(rougher) — симметричное среднее абсолютное процентное отклонение после этапа флотации;
* sMAPE(final) — симметричное среднее абсолютное процентное отклонение после проведения всех этапов очистки.

```python
def s_mape_final(rougher, final):
    
    return 0.25 * rougher + 0.75 * final
```

### Обучение моделей.

Так как целевых признакаа два, разделим обучение на два этапа - предсказание значений после флотации и предсказание значений финальной обработки. На каждом этапе обучим модели отдельно. Для повышения эффективности обучения воспользуемся методом кросс-валидации.

Для обучения выберем следующие модели регрессий:
- линейная
- решающее дерево
- случайный лес решающих деревьев

Предварительно проведем выделение целевого признака для каждой подбрки данных. Также уберем из данных lkz предсказаний этапа флотации информацию о последующих обработках.

```python
# функция для расчета среднего откланения предсказания
def cross_valid(features, target, model):
    
    pipe = make_pipeline(StandardScaler(), model)
       
    scores = cross_val_score(pipe, features, 
                             target, cv=7, scoring=scorer) 
    
    final_score = (scores * (-1)).mean()
    
    return final_score
```

#### Регрессия решающго дерева.

Решающее дерево - еще одна из простых моделей предсказаний. Посмотрим, будут ли показатели решающего дерева лучше, чем у линейной регрессии.

```python
def dtr(features, target, titl):
    
    best_model = None
    best_scores = 100
    best_depth = 0
    
    for depth in range(5, 16):
    
        model_dtr = DecisionTreeRegressor(random_state=12345,
                                          max_depth = depth)
    
        scores_dtr = cross_valid(features, target, model_dtr)
    
        if scores_dtr < best_scores:
            best_scores = scores_dtr
            best_model = model_dtr
            best_depth = depth
            
    print('Лучшая модель решаюшего дерева достигнута' +
          ' при глубине дерева', best_depth, '.')
    print('На', titl,'sMAPE модели составляет',
          '%.2f' %best_scores ,'%.')
    
    return best_scores, best_model
```

```python
rougher_dtr, rougher_model_dtr = \
                        dtr(features_rougher, target_rougher,
                            'этапе флотации')
```

```python
final_dtr, final_model_dtr = dtr(features_final, target_final,
                                 'финальном этапе')
```

```python
end_dtr = s_mape_final(rougher_dtr, final_dtr)
print('sMAPE всей модели составляет','%.2f' %end_dtr ,'%.')
```

На этапе флотации модель решающего дерева показывает результат немного хуже, чем модель линейной регрессии(6,28% против 5,75%) А на предсказании финального результата наоборот, показывает показатели лучше(8,6% против 10.46%). На данном этапе исследования, рекомендовано использовать для предсказания на этапе флотации модель линейной регрессии, а для прогноза финального этапа обработки - модель решающего дерева. 

Попробуем еще улучшить показатели, для этого перейдем к более сложной модели регрессии.

#### Случайный лес решающих деревьев.

Продолжим поиск лучшей модели предсказаний со случайным лесом решающих деревьев. Поиск лучшей модели может занять некоторое время(7,5 минут для этапа флотации и 17 минут для финального этапа).

```python
def rtr(features, target, titl):
    
    best_model = None
    best_scores = 100
    best_depth = 0
    best_n_est = 0

    for est in range(100, 121, 10):
        
        for depth in range(4, 8):
    
            model_rtr = RandomForestRegressor(random_state=12345, 
                                              n_estimators=est, 
                                              max_depth=depth)
    
            scores_rtr = cross_valid(features, target, model_rtr)

            if scores_rtr < best_scores:
                best_scores = scores_rtr
                best_model = model_rtr
                best_depth = depth
                best_n_est = est
                        
    print('Лучшая модель случайного леса достигнута при', best_n_est, 
                                   'деревьев и глубине', best_depth, '.')
    print('На', titl,'sMAPE модели составляет','%.2f' %best_scores ,'%.')
    return best_scores, best_model
```

```python
rougher_rtr, rougher_model_rtr = \
                rtr(features_rougher, target_rougher, 'этапе флотации')
```

```python
final_rtr, final_model_rtr = rtr(features_final,
                                 target_final,
                                 'финальном этапе')
```

```python
end_rtr = s_mape_final(rougher_rtr, final_rtr)
print('sMAPE всей модели составляет','%.2f' %end_rtr ,'%.')
```

#### Лучшая модель.

По результатам исследования для предсказаний результатов флотации для предсказания возьмем модель линейной регрессии, а для предсказаний результатов финальной очистки - слечайный лес решающих деревьев. Лучшие показатели модели решающего леса были при параметрах: число деревьев 100, а максимальная глубина 6.

```python
# функция обучения модели
def model_training(features, target, model):

    # обучение модели
    model = model.fit(features, target) 
    
    predict = model.predict(features)
    
    s = s_mape(target, predict)
    
    print('Модель обучена. На обучающих данных sMAPE =', '%.2f' %s,'%.')
    
    return model, s
```
